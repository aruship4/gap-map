```python
# Convert knowledge columns to numeric
df["Ln"] = pd.to_numeric(df["Ln"], errors="coerce")
df["Ln-1"] = pd.to_numeric(df["Ln-1"], errors="coerce")

# Now this will work
df["learningGain"] = df["Ln"] - df["Ln-1"]
skill_summary = (
    df.groupby("skill")
    .agg(
        attempts=("correct","count"),
        accuracy=("correct","mean"),
        avg_attempts=("attemptCount","mean"),
        avg_hints=("hintCount","mean"),
        avg_time=("timeTaken","mean"),
        learning_gain=("learningGain","mean"),
        avg_consecutive_errors=("consecutiveErrorsInRow","mean"),
        bored=("RES_BORED","mean"),
        confused=("RES_CONFUSED","mean"),
        frustrated=("RES_FRUSTRATED","mean"),
        offtask=("RES_OFFTASK","mean"),
        gaming=("RES_GAMING","mean")
    )
)

skill_summary["failure_rate"] = 1 - skill_summary["accuracy"]

display(skill_summary.head())

skill_summary.to_csv("skill_summary.csv")
```

```python
top_failures = (
    skill_summary
    .sort_values("failure_rate", ascending=False)
    .head(20)
)

display(top_failures)

top_failures.sort_values("failure_rate")["failure_rate"].plot.barh()

plt.title("Highest Failure Rate Skills")

plt.xlabel("Failure Rate")

plt.show()
```

```python
top_hints = (
    skill_summary
    .sort_values("avg_hints", ascending=False)
    .head(20)
)

display(top_hints)

top_hints.sort_values("avg_hints")["avg_hints"].plot.barh()

plt.title("Average Hint Usage")

plt.show()
```

```python
top_time = (
    skill_summary
    .sort_values("avg_time", ascending=False)
    .head(20)
)

display(top_time)

top_time.sort_values("avg_time")["avg_time"].plot.barh()

plt.title("Average Response Time")

plt.show()
```

```python
df["struggleScore"] = (
    df["attemptCount"]
    + df["hintCount"]
    + df["consecutiveErrorsInRow"]
)

struggle_summary = (
    df.groupby("skill")
    ["struggleScore"]
    .mean()
    .sort_values(ascending=False)
)

display(struggle_summary.head(20))

struggle_summary.head(20).sort_values().plot.barh()

plt.title("Highest Average Struggle Score")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Columns to include
corr_cols = [
    "correct",
    "attemptCount",
    "hintCount",
    "timeTaken",
    "consecutiveErrorsInRow",
    "frPast5WrongCount",
    "frPast8WrongCount",
    "Ln-1",
    "Ln",
    "learningGain",
    "RES_BORED",
    "RES_CONFUSED",
    "RES_FRUSTRATED",
    "RES_OFFTASK",
    "RES_GAMING"
]

# Ensure they're numeric
for col in corr_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Compute correlation matrix
corr = df[corr_cols].corr()

# Plot
fig, ax = plt.subplots(figsize=(12, 10))

im = ax.imshow(corr, cmap="coolwarm", vmin=-1, vmax=1)

# Tick labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))

ax.set_xticklabels(corr.columns, rotation=90)
ax.set_yticklabels(corr.columns)

# Add correlation values
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        ax.text(
            j,
            i,
            f"{corr.iloc[i, j]:.2f}",
            ha="center",
            va="center",
            fontsize=8,
            color="black"
        )

plt.title("Correlation Matrix")

plt.colorbar(im, label="Correlation")

plt.tight_layout()

plt.show()
```
